```python
import cv2
import numpy as np
```

```python
img=cv2.imread('/content/WhatsApp Image 2026-03-23 at 9.13.53 AM.jpeg',0)
```

```python
gauss=img + np.random.normal(0,60,img.shape)
```

```python
print(gauss)
```

```python
gauss=np.clip(gauss,0,255).astype(np.uint8)
```

```python
print(gauss)
```

```python
sp=img.copy()
```

```python
prob=0.02
rand=np.random.rand(*img.shape)
sp[rand<prob]=0
```

```python
sp[rand>1-prob]=255
```

```python
print(rand)
```

```python
cv2.imwrite("gauss.jpg",gauss)
```
