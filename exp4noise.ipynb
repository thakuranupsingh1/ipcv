```python
img=cv2.imread('/content/noisy.ppm.png',0)
```

```python
f = np.fft.fft2(img)
```

```python
fshift=np.fft.fftshift(f)
```

```python
row,cols=img.shape
```

```python
crow,ccol=row//2,cols//2
```

```python
mask_lp=np.zeros((row,cols),np.uint8)
```

```python
mask_lp[crow-30:crow+30,ccol-30:ccol+30]=1
```

```python
lp=np.fft.ifft2(np.fft.ifftshift(fshift*mask_lp))
```

```python
cv2.imwrite("lpf_noisy.ppm.png",np.abs(lp))
```
